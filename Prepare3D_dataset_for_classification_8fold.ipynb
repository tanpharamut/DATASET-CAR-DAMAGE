# ðŸ“ Created Data ---> CSV. file 

## Get Image path all directory (8 id) 

```python
import glob

path_root =  '/media/SSD/Data_photogram_car/'

root_path = []
img_pathID_ = []

dir_a = os.listdir(path_root)  
dir_a
for a in range(len(dir_a)):
    img_pathID = []
    aa = path_root+dir_a[a]+'/'
    dir_b = os.listdir(aa)
    for b in range(len(dir_b)):
        bb = aa+dir_b[b]+'/'
        dir_c = os.listdir(bb)
        for c in range(len(dir_c)):
            cc = bb+dir_c[c]+'/'
            dir_d = os.listdir(cc)
            for d in range(len(dir_d)):
                dd = cc+dir_d[d]+'/'
                dir_e = os.listdir(dd)
                for e in range(len(dir_e)):
                    ee = dd+dir_e[e]+'/'
                    root_path.append(ee)
                    files_img = glob.glob(f"{ee}*.png")
                    for f in files_img:
                        img_pathID.append(f)
    img_pathID_.append(img_pathID)
    #img_pathID_.append(img_pathID)
```

```python
flat_files_ = []
for sublist in img_pathID_:
    for item in sublist:
        flat_files_.append(item)
print(f"à¸ˆà¸³à¸™à¸§à¸™à¸ à¸²à¸žà¸—à¸±à¹‰à¸‡à¸«à¸¡à¸” : --->> {len(flat_files_)}")
flat_files_.sort()
flat_files_
```

### ðŸŒ€Save to CSV.

```python
id_, dam_, P_,dis_,ang_,fileName_ = [], [], [],[], [], [] 

for file in flat_files_:
    id0 = file.split('/')[4] 
    id_.append(id0)
    #print(id0)
    dam = file.split('/')[5] 
    dam_.append(dam)
    P = file.split('/')[6] 
    P_.append(P)
    dis = file.split('/')[7]
    #print(dis)
    dis_.append(dis)
    ang = dis = file.split('/')[8] 
    #print(ang)
    ang_.append(ang)
    fileName = file.split('/')[-1] 
    #print(fileName)
    fileName_.append(fileName)

Table_path = {'ID':id_,'Damage':dam_,'p':P_, 'FileName':fileName_,'distance':dis_, 'Angle': ang_, 'img_path':flat_files_}
Table_path_ = pd.DataFrame(Table_path, columns=['ID','Damage','p','FileName', 'distance', 'Angle', 'img_path'])
print(Table_path_.shape)
Table_path_.head()
```

```python
set_f = ['id2', 'id6', 'id8', 'id10', 'id12', 'id13', 'id14', 'id26']
fold_ = []
for i in range(len(Table_path_)):
    f = Table_path_['ID'][i]
    #print(f)
    for j, k in zip(set_f, range(1,9)):
        #print(j)
        if f == j:
            #print(k)
            fold_.append(f'fold-{k}')
print(len(fold_))
fold_
```

```python
## Add Columns fold
Table_path_['Fold'] = fold_
print(Table_path_.shape)
Table_path_.head()
```

# ðŸ“ Created Folder train

- ðŸš¨ path to save: ``/media/SSD/Data_photogram_Pcar/8-Fold``

```python
import shutil as sh
import time

path2save = '/media/SSD/Data_photogram_Pcar/8-Fold/'
fold0 = list(set(Table_path_['Fold']))
fold0.sort()
for i in range(len(fold0)):
    df0 = Table_path_[Table_path_['Fold'] == fold0[i]]
    path_img = df0['img_path'].tolist()
    name_img = df0['FileName'].tolist()
    if not os.path.exists('{}'.format(path2save)) :
             os.makedirs('{}'.format(path2save))
    for j in range(len(path_img)):
        sh.copy("{}".format(path_img[j]),"{}{}".format(path2save, name_img[j]))
    print(f'On process Copy at fold -->> [ {fold0[i]} ] ')
    print('-'*100)
    time.sleep(5)    
```
