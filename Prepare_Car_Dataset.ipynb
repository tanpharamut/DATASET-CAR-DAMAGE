## üìÅFunction Generate Car DataSet (2)

> ### (‡πÅ‡∏ö‡∏ö‡∏ó‡∏µ‡πà 1 : Path image ‡∏à‡∏≤‡∏Å‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏£‡∏∏‡∏ô‡πÅ‡∏£‡∏á‡πÄ‡∏õ‡πá‡∏ô image ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô)

> - Return (CSV. file every 'p1,p2,...' )
> - CSV. file contain data 3 Batch
> - In each batch is images path from same 'ID' 

```python
### Function (Table_id2p1_all==DataFrame)
import random

def GeneDatabath(Table_path_, ID): ##* Table_path_; DataFrame ,ID;String
    Table_id = Table_path_[Table_path_['ID'] == ID].reset_index(drop=True)
    path_save = f'/media/SSD/Data_photogram_Pcar/DiffBatch/{ID}/'
    print(f'Path To Save Car DataSet ---> [ {path_save} ] ')
    print('-'*100)

    if not os.path.exists(path_save) :
        os.makedirs(path_save)

    set_dam = ['Base'  ,'Minor',  'Moderate'  ,'Severe']
    set_p = list(set(Table_id['p']))
    set_p.sort()

    for p in range(len(set_p)):
        Table_idp = Table_id[Table_id['p']==set_p[p]]
        Table_idp_all = []
        for d in set_dam:
            Table_d = Table_idp[Table_idp['Damage']==d].reset_index(drop=True)
            Table_idp_all.append(Table_d)

        Table_base = Table_idp_all[0]
        Table_Minor = Table_idp_all[1]
        Table_Moderate = Table_idp_all[2]
        Table_Severe = Table_idp_all[3]
        list_base1 = Table_base['img_path'].tolist()
        list_base2 = Table_base['img_path'].tolist()
        list_base3 = Table_base['img_path'].tolist()
        list_Minor = Table_Minor['img_path'].tolist()
        list_Moderate  = Table_Moderate['img_path'].tolist()
        list_Severe = Table_Severe['img_path'].tolist()

        databatch_  = []

        for i in range(len(list_base1)):
            random_bath = []

            random_base1 = random.choice(list_base1)
            #list_base1.remove(random_base1)
            #try:
            list_base2.remove(random_base1)
            #except:
            if len(list_base2) == 0:
                list_base3.remove(random_base1)
                random_base2 = random.choice(list_base3)
            else:
                random_base2 = random.choice(list_base2)
            random_Minor = random.choice(list_Minor)
            list_Minor.remove(random_Minor)
            random_Moderate = random.choice(list_Moderate)
            list_Moderate.remove(random_Moderate)
            random_Severe = random.choice(list_Severe)
            list_Severe.remove(random_Severe)
            list_base1.remove(random_base1)

            random_bath1 = [random_base1, random_base2,random_Minor]
            random_bath2 = [random_base1, random_Minor,random_Moderate]
            random_bath3 = [random_base1, random_Moderate,random_Severe]
            #andom_bath1.append()
            random_bath.append(random_bath1)
            random_bath.append(random_bath2)
            random_bath.append(random_bath3)
            #random_bath1.append(random_base1)

            databatch = pd.DataFrame(list(zip(random_bath[0], random_bath[1], random_bath[2])),
                           columns =['batch1', 'batch2', 'batch3'])
            databatch.to_csv(f'{path_save}car{ID}_{set_p[p]}_batch{i+1}.csv')
            databatch_.append(databatch)

        print(f'Save Data Batch SET {ID}---> ({p+1}). {set_p[p]} -->>> len [ {len(databatch_)} ]')
    print('*'*120)
```

## üìÅFunction Generate Car DataSet (2)

> ### (‡πÅ‡∏ö‡∏ö‡∏ó‡∏µ‡πà 2 : Path image ‡∏à‡∏≤‡∏Å‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏£‡∏∏‡∏ô‡πÅ‡∏£‡∏á‡πÑ‡∏°‡πà‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡πâ‡∏≠‡∏á image ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô/‡∏¢‡∏Å‡πÄ‡∏ß‡πâ‡∏ô Base ‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡πÉ‡∏ô batch ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô)

> - Return (CSV. file every 'p1,p2,...' )
> - CSV. file contain data 3 Batch
> - In each batch is images path from same 'ID' 

```python
### Function (Table_id2p1_all==DataFrame)
import random

def GeneDatabath_random(Table_path_, ID): ##* Table_path_; DataFrame ,ID;String
    Table_id = Table_path_[Table_path_['ID'] == ID].reset_index(drop=True)
    path_save = f'/media/SSD/Data_photogram_Pcar/RandomBatch/{ID}/'
    print(f'Path To Save Car DataSet ---> [ {path_save} ] ')
    print('-'*100)

    if not os.path.exists(path_save) :
        os.makedirs(path_save)

    set_dam = ['Base'  ,'Minor',  'Moderate'  ,'Severe']
    set_p = list(set(Table_id['p']))
    set_p.sort()

    for p in range(len(set_p)):
        Table_idp = Table_id[Table_id['p']==set_p[p]]
        Table_idp_all = []
        for d in set_dam:
            Table_d = Table_idp[Table_idp['Damage']==d].reset_index(drop=True)
            Table_idp_all.append(Table_d)

        Table_base = Table_idp_all[0]
        Table_Minor = Table_idp_all[1]
        Table_Moderate = Table_idp_all[2]
        Table_Severe = Table_idp_all[3]
        ## Create list of data in each p1,p2,p3,....
        list_base1 = Table_base['img_path'].tolist()
        list_base2 = Table_base['img_path'].tolist()
        list_base2_5 = Table_base['img_path'].tolist()
        list_base3 = Table_base['img_path'].tolist()
        list_base4 = Table_base['img_path'].tolist()
        list_Minor1 = Table_Minor['img_path'].tolist()
        list_Minor2 =  Table_Minor['img_path'].tolist()
        list_Moderate1 = Table_Moderate['img_path'].tolist()
        list_Moderate2 = Table_Moderate['img_path'].tolist()
        list_Severe = Table_Severe['img_path'].tolist()
        
        databatch_  = []
        for i in range(len(list_base1)):
            random_bath = []

            random_base1 = random.choice(list_base1)
            list_base2.remove(random_base1)

            if len(list_base2) == 0:
                list_base2_5.remove(random_base1)
                random_base2 = random.choice(list_base2_5)
            else:
                random_base2 = random.choice(list_base2)

            random_base3 = random.choice(list_base3)
            random_base4 = random.choice(list_base4)

            random_Moderate1 = random.choice(list_Moderate1)
            random_Minor1 = random.choice(list_Minor1)

            random_Minor2 = random.choice(list_Minor2)
            random_Moderate2 = random.choice(list_Moderate2)

            random_Severe = random.choice(list_Severe)
            list_Severe.remove(random_Severe)

            list_base1.remove(random_base1) ##**
            list_base3.remove(random_base3) ##**
            list_base4.remove(random_base4) ##***
            list_Minor1.remove(random_Minor1) ##**
            list_Minor2.remove(random_Minor2)  ##**
            list_Moderate1.remove(random_Moderate1) ##**
            list_Moderate2.remove(random_Moderate2)

            random_bath1 = [random_base1, random_base2, random_Minor1]
            random_bath2 = [random_base3, random_Minor2, random_Moderate1]
            random_bath3 = [random_base4, random_Moderate2, random_Severe]
            random_bath.append(random_bath1)
            random_bath.append(random_bath2)
            random_bath.append(random_bath3)
            databatch = pd.DataFrame(list(zip(random_bath[0], random_bath[1], random_bath[2])),
                           columns =['batch1', 'batch2', 'batch3'])
            databatch.to_csv(f'{path_save}car{ID}_{set_p[p]}_batch{i+1}.csv')
            databatch_.append(databatch)

        print(f'Save Data Batch SET {ID}---> ({p+1}). {set_p[p]} -->>> len [ {len(databatch_)} ]')
    print('*'*120)
```
